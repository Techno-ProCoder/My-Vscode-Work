```python
# =============================
# 1. MATH LIBRARY
# =============================

import math

# Square root
print("Square root of 16:", math.sqrt(16))

# Power
print("2 raised to power 3:", math.pow(2, 3))

# Trigonometric functions (angles in radians)
print("sin(0):", math.sin(0))
print("cos(0):", math.cos(0))

# Constants
print("Value of pi:", math.pi)
print("Value of e:", math.e)

# Rounding
print("Floor of 4.7:", math.floor(4.7))
print("Ceil of 4.1:", math.ceil(4.1))
```

```python
# =============================
# 2. NUMPY LIBRARY
# =============================

import numpy as np

# Create arrays
arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)

# Basic statistics
print("Mean:", np.mean(arr))
print("Standard Deviation:", np.std(arr))
print("Sum:", np.sum(arr))
print("Maximum:", np.max(arr))
print("Minimum:", np.min(arr))

# Array operations
print("Array squared:", np.square(arr))
print("Array + 5:", arr + 5)

# Special arrays
print("Zeros array:", np.zeros((2, 3)))
print("Ones array:", np.ones((3, 2)))
print("Range of numbers:", np.arange(0, 10, 2))
```

```python
# =============================
# 3. PANDAS LIBRARY
# =============================

import pandas as pd

# Creating a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Score': [85, 90, 95, 88]
}
df = pd.DataFrame(data)
print("DataFrame:\n", df)

# Viewing basic info
print("\nHead of DataFrame:\n", df.head(2))
print("\nDescription of DataFrame:\n", df.describe())

# Selecting a column
print("\nAges:\n", df['Age'])

# Filtering data
print("\nPeople with Score > 88:\n", df[df['Score'] > 88])

# Adding a new column
df['Passed'] = df['Score'] >= 90
print("\nDataFrame with 'Passed' column:\n", df)

# Sorting
print("\nSorted by Age:\n", df.sort_values(by='Age', ascending=False))
```
