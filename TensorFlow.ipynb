```python
"""
Lecture: TensorFlow Basics
This file demonstrates core TensorFlow concepts step-by-step.
"""

# ======================================================
# 1. Importing TensorFlow
# ======================================================

import tensorflow as tf
import numpy as np

print("TensorFlow version:", tf.__version__)

# ======================================================
# 2. What is TensorFlow?
# ======================================================

# TensorFlow is an open-source library for:
# - Numerical computation
# - Deep Learning
# - Neural Networks
print("TensorFlow is used for deep learning and numerical computation")

# ======================================================
# 3. Tensors (Core Data Structure)
# ======================================================

print("\nTENSOR CREATION")

# Creating tensors from Python lists
a = tf.constant([1, 2, 3])
b = tf.constant([[1, 2], [3, 4]])

print("Tensor a:", a)
print("Tensor b:\n", b)

# Rank (dimensions)
print("Rank of a:", tf.rank(a))
print("Shape of b:", b.shape)

# ======================================================
# 4. Tensor Data Types
# ======================================================

x = tf.constant([1, 2, 3], dtype=tf.float32)
y = tf.constant([4, 5, 6], dtype=tf.int32)

print("x dtype:", x.dtype)
print("y dtype:", y.dtype)

# ======================================================
# 5. Tensor Operations
# ======================================================

print("\nTENSOR OPERATIONS")

# Element-wise operations
print("Addition:", tf.add(x, tf.cast(y, tf.float32)))
print("Multiplication:", tf.multiply(x, tf.cast(y, tf.float32)))

# Matrix multiplication
m1 = tf.constant([[1, 2], [3, 4]])
m2 = tf.constant([[5, 6], [7, 8]])

print("Matrix multiplication:\n", tf.matmul(m1, m2))

# ======================================================
# 6. Conversion Between NumPy & TensorFlow
# ======================================================

print("\nNUMPY ↔ TENSORFLOW")

np_arr = np.array([10, 20, 30])
tf_tensor = tf.convert_to_tensor(np_arr)

print("NumPy array:", np_arr)
print("Converted to tensor:", tf_tensor)

print("Back to NumPy:", tf_tensor.numpy())

# ======================================================
# 7. Variables (Trainable Parameters)
# ======================================================

print("\nVARIABLES")

# Variables are mutable tensors
w = tf.Variable([1.0, 2.0, 3.0])
print("Initial variable:", w)

w.assign_add([1.0, 1.0, 1.0])
print("After update:", w)

# ======================================================
# 8. Automatic Differentiation (Gradient Tape)
# ======================================================

print("\nAUTOMATIC DIFFERENTIATION")

x = tf.Variable(3.0)

with tf.GradientTape() as tape:
    y = x ** 2

dy_dx = tape.gradient(y, x)
print("dy/dx at x=3:", dy_dx)

# ======================================================
# 9. Simple Neural Network (Keras API)
# ======================================================

print("\nSIMPLE NEURAL NETWORK")

# Creating a simple model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(8, activation='relu', input_shape=(4,)),
    tf.keras.layers.Dense(1)
])

model.summary()

# ======================================================
# 10. Compiling the Model
# ======================================================

model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)

print("Model compiled successfully")

# ======================================================
# 11. Dummy Training Example
# ======================================================

# Dummy data
X = np.random.random((10, 4))
y = np.random.random((10, 1))

# Training
model.fit(X, y, epochs=3, verbose=1)

# ======================================================
# End of Lecture
# ======================================================

print("\n✅ END OF TENSORFLOW INTRODUCTION")
```
